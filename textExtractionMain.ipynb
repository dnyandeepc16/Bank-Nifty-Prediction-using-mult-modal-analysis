```python
from finbert_embedding.embedding import FinbertEmbedding
import numpy as np
import docx
import os
import pickle
```

```python
truepath = os.getcwd()
```

```python
def getText(filename):
    doc = docx.Document(filename)
    fullText = []
    for para in doc.paragraphs:
        fullText.append(para.text)
    return '\n'.join(fullText)
```

```python
finbert = FinbertEmbedding()
```

```python
sentence_embedding_list_final = []

for i in os.listdir("Dataset/"):
    textDir = "Dataset/"+i+"/texts/"
    os.chdir(textDir)
    
    text_tokens_List = []
    word_embeddings_List =  []
    sentence_embedding_List_temp = []

    for j in os.listdir():
        text = getText(j)

        token = finbert.tokens
        word_embeddings = finbert.word_vector(text)
        sentence_embedding = finbert.sentence_vector(text)

        #print("Text Tokens: ", finbert.tokens)
        #print ('Shape of Word Embeddings: %d x %d' % (len(word_embeddings), len(word_embeddings[0])))
        #print("Shape of Sentence Embedding = ",len(sentence_embedding))

        text_tokens_List.append(token)
        word_embeddings_List.append(word_embeddings)
        sentence_embedding_List.append(sentence_embedding)
        
    #sentence_embedding_list_final.append(sentence_embedding_List)
    print(i + " Feature Extraction Done!")
    
    os.chdir(truepath)
```
